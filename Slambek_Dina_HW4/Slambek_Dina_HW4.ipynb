```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
day = pd.read_csv("day.csv")
day.shape
```

Dataset characteristics	
-day.csv - bike sharing counts aggregated on daily basis. Records: 731 days
day.csv have the following fields:
	- instant: record index
	- dteday : date
	- season : season (1:springer, 2:summer, 3:fall, 4:winter)
	- yr : year (0: 2011, 1:2012)
	- mnth : month ( 1 to 12)
	- holiday : weather day is holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
	- weekday : day of the week
	- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
	+ weathersit : 
		- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
		- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
		- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
		- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
	- temp : Normalized temperature in Celsius. The values are divided to 41 (max)
	- atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)
	- hum: Normalized humidity. The values are divided to 100 (max)
	- windspeed: Normalized wind speed. The values are divided to 67 (max)
	- casual: count of casual users
	- registered: count of registered users
	- cnt: count of total rental bikes including both casual and registered

```python
#your code
fig = plt.figure()

plt.rcParams['figure.figsize'] = [10,8]


fig.suptitle('Bike Rentals at Different Temperatures By Season', fontsize=14, fontweight="bold", position=(0.5,1.01))

ax1 = fig.add_subplot(221)
ax1.scatter('temp', 'cnt', data=day[day['season']==1], color="green", marker="o", linestyle="None")
ax1.set_title('Spring', fontdict={"size":12}, fontweight="bold", color="green")
ax1.set_ylabel("Count of Total Rental Bikes", fontdict={"size":12}, fontweight="bold", position=(0,-0.1))
ax1.grid()


ax2 = fig.add_subplot(222)
ax2.scatter('temp', 'cnt', data=day[day['season']==2], color="orange", marker="o", linestyle="None")
ax2.set_title('Summer', fontdict={"size":12}, fontweight="bold", color="orange")
ax2.grid()


ax3 = fig.add_subplot(223)
ax3.scatter('temp', 'cnt', data=day[day['season']==3], color="brown", marker="o", linestyle="None")
ax3.set_title('Fall or Autumn', fontdict={"size":12}, fontweight="bold", color="brown")
ax3.grid()

ax4 = fig.add_subplot(224)
ax4.scatter('temp', 'cnt', data=day[day['season']==4], color="blue", marker="o", linestyle="None")
ax4.set_title('Winter', fontdict={"size":12}, fontweight="bold", color="blue")
ax4.set_xlabel("Normalized temperature", fontdict={"size":12}, fontweight="bold", position=(-0.1,0));
ax4.grid()
```

- Figsize = (8, 7);
- Palette: green, orange, brown, blue;

```python
#your code
plt.rcParams['figure.figsize'] = [8, 7]

spring = plt.scatter('temp', 'cnt', data=day[day['season']==1], marker='.', linestyle="None", color='green')
summer = plt.scatter('temp', 'cnt', data=day[day['season']==2], marker='x', linestyle="None", color='orange')
autumn = plt.scatter('temp', 'cnt', data=day[day['season']==3], marker='s', linestyle="None", color='brown')
winter = plt.scatter('temp', 'cnt', data=day[day['season']==4], marker='+', linestyle="None", color='blue')
plt.legend(handles=(spring,summer,autumn,winter),
           labels=('1', '2', '3', '4'),
           title="season", title_fontsize=10,
           scatterpoints=1,
           bbox_to_anchor=(0, 1), loc=2,
           ncol=1,
           fontsize=12)
plt.title('Bike Rentals at Different Temperature By Season', fontdict={"size":16}, fontweight="bold", color="black")
plt.xlabel("Normalized temperature",fontdict={"size":14}, fontweight="bold")
plt.ylabel("Count of Total Rental Bikes", fontdict={"size":14}, fontweight="bold")
```
